```python
df_duplicated = df[df.duplicated()]
if not df_duplicated.empty:
    print("Duplicate rows found:")
    df = df.drop_duplicates()
df.info()
```

### Task 3 & 4 :
### Exploratory Data Analysis (EDA) & Data Visualization

```python
print("Basic Statistics")
df.describe()
```

```python
print("Revenue Product")
revenue_product = df.groupby('Product')['TotalPrice'].sum().sort_values(ascending=False)

df["YearMonth"]= df["Date"].dt.to_period('M')
print("Monthly Order Count")
df.groupby("YearMonth")["OrderID"].count()
```

```python
numeric_cols = df.select_dtypes(include="number").columns.tolist()
print("Outliers per column (IQR method):")
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1 - 1.5 * IQR) | (df[col] > Q3 + 1.5 * IQR)]
df[numeric_cols].plot(kind="box", subplots=True, layout=(1, 4), figsize=(14, 5))
plt.suptitle("Outlier Detection — Boxplots")
plt.tight_layout()
plt.show()
```

```python
# Summary of EDA findings
print("=" * 45)
print("EDA SUMMARY")
print("=" * 45)
print(f"Total orders      : {len(df):,}")
print(f"Total revenue     : ${df['TotalPrice'].sum():,.2f}")
print(f"Avg order value   : ${df['TotalPrice'].mean():,.2f}")
print(f"Most ordered item : {df['Product'].value_counts().idxmax()}")
print(f"Top revenue item  : {df.groupby('Product')['TotalPrice'].sum().idxmax()}")
print(f"Cancellation rate : {(df['OrderStatus'] == 'Cancelled').mean():.1%}")
```
